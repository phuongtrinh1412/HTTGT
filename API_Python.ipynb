```python
from bs4 import BeautifulSoup
import urllib.request
import csv

url =  'https://stackoverflow.com/'
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, 'html.parser')



objs = soup.find_all('div', class_='question-summary narrow')

with open('data.txt', 'a') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerow(['Votes', 'Answer', 'Question Id', 'Question'])
    csvFile.close()
    
for element in objs:
    data = []
    status = element.find_all('div', class_='mini-counts')
    for e in status:
        news = e.find('span')
        title = news.get('title')
        data.append(title)
    
    question_link = element.find('a')
    link = question_link.get('href')
    contentArr = link.split('/')
    data.append(contentArr[2]) 
    data.append(contentArr[3])
    
    with open('data.txt', 'a') as csvFile:
        writer = csv.writer(csvFile)
        writer.writerow(data)
    csvFile.close()

import pandas as pd
name = "data.txt"
data = pd.read_csv(name, lineterminator='\n')
data.head(data.size)
```
